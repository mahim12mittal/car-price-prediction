```python
X = ev.drop(columns = ['Brand','Model','RapidCharge','PlugType','Price $'])   #independent variables 
Y = ev['Price $'] # dependent
```

```python
X
```

```python
Y
```

```python
X_nor = pd.DataFrame(X,columns=['AccelSec','TopSpeed_KmH','Range_Km','Efficiency_WhKm','FastCharge_KmH','Seats'])
```

```python
X_nor
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X_nor,Y,test_size=0.4,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor  
lrmodel = KNeighborsRegressor()
```

```python
lrmodel
```

```python
lrmodel.fit(X_train,Y_train)
```

```python
acc = lrmodel.score(X_test,Y_test)
```

```python
acc
```

```python
import pickle
pickle.dump(lrmodel,open('knncartest_model','wb'))
```

```python
model = pickle.load(open('knncartest_model','rb'))
```

```python
result = model.predict(X_test)
```
